```python
from IPython.display import clear_output
def display(arr):
        
    print("    |    |    ")
    print(f'  {arr[7]} |  {arr[8]} | {arr[9]} ')
    print("    |    |    ")
    print("----------------")
    print("    |    |    ")
    print(f'  {arr[4]} |  {arr[5]} | {arr[6]} ')
    print("    |    |    ")
    print("----------------")
    print("    |    |    ")
    print(f'  {arr[1]} | {arr[2]}  | {arr[3]} ')
    print("    |    |    ")
    
    
```

```python
def full_check(arr):
    for i in range(1,9):
        if(arr[i]==' '):
            return False
        
    return True
            
```

```python
def position_marker(arr,ans,turn):
    sim=int(ans)
    cha=''
    if(turn == 1):
        cha='X'
        arr[sim]=cha
    else:
        cha='O'
        arr[sim]=cha    
```

```python
def check_game(arr):
    if(arr[1]==arr[2]==arr[3]!=' '):
        return 'Win'
    elif(arr[4]==arr[5]==arr[6]!=' '):
        return 'Win'
    elif(arr[7]==arr[8]==arr[9]!=' '):
        return 'Win'
    elif(arr[1]==arr[5]==arr[9]!=' '):
        return 'Win'
    elif(arr[7]==arr[5]==arr[3]!=' '):
        return 'Win'
    elif(arr[1]==arr[4]==arr[7]!=' '):
        return 'Win'
    elif(arr[2]==arr[5]==arr[8]!=' '):
        return 'Win'
    elif(arr[3]==arr[6]==arr[9]!=' '):
        return 'Win'
```
